```python
df=df.dropna()
```

```python
df['column']=df['CustomerID'].fillna(df['CustomerID'].mean())
```

```python
df['column']=df['Age'].fillna(df['Age'].mean())
```

```python
df['column']=df['Annual Income (k$)'].fillna(df['Annual Income (k$)'].mean())
```

```python
df['column']=df['Spending Score (1-100)'].fillna(df['Spending Score (1-100)'].mean())
```

```python
df['gender']=df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df = df.drop_duplicates()
```

```python
df['gender']=df['gender'].str.lower()
```

```python
df.columns=df.columns.str.strip().str.lower().str.replace(" ","_")
```

```python
df.to_csv("cleaned_dataset.csv",index=False)
```
